```python
import pandas as pd
import numpy as py

import copy #need it to do deep copy instead of copy() function

import matplotlib as plt
import matplotlib.pyplot as plt #need it for the bar plots
import seaborn as sns

#for converting the lat&lon to Km
from geopy import Point
from geopy.distance import distance

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")

#to plot the lat&lon in map plot
import cartopy.crs as crs
import cartopy.io.img_tiles as cimgt
import io
from urllib.request import urlopen, Request
from PIL import Image

def image_spoof(self, tile): 
    url = self._image_url(tile) 
    req = Request(url) 
    req.add_header('User-agent','Anaconda 3') 
    fh = urlopen(req) 
    im_data = io.BytesIO(fh.read())
    fh.close() 
    img = Image.open(im_data)
    img = img.convert(self.desired_tile_form)
    return img, self.tileextent(tile), 'lower'
```

## Data Preprocessing

```python
#the total Nan values is considered 4% from the full data, so will drop them all

bike_df = bike_df.dropna(axis=0, how='any', inplace=False)
```

```python
bike_df.isnull().sum()
```

>Because the date and time are combined in same column ( for [Start-Time] and [End_Time] ), will extract the useful data only. but before extraction step, we need to converted them into datatime ( bec. they are currently in object type )

References:

1- https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html


2- https://towardsdatascience.com/mastering-dates-and-timestamps-in-pandas-and-python-in-general-5b8c6edcc50c 


3- https://stackoverflow.com/questions/54733465/pandas-how-to-extract-hhmm-from-datetime-column-in-python/54733582

4- https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.weekday.html

5- https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.month.html

```python
#convert from object to datetime

bike_df['start_time'] = pd.to_datetime(bike_df['start_time'])
bike_df['end_time'] = pd.to_datetime(bike_df['end_time'])

bike_df.dtypes
```

```python
#extract the month

bike_df['month'] = bike_df['start_time'].dt.month
bike_df.head(3)
```

```python
#extract the weekday

bike_df['weekday'] = bike_df['start_time'].dt.weekday
bike_df.head(3)
```

```python
#convert (duration_sec) to (hours)

bike_df['hours'] = bike_df['duration_sec']/3600
bike_df.head(3)
```

>need to extract the KiloMeter data from Lan and Lon datasets

References:

1- https://pypi.org/project/geopy/

2- https://stackoverflow.com/questions/55909305/using-geopy-in-a-dataframe-to-get-distances

```python
bike_df
```

```python
#first combine then into columns of points

bike_df['start_latlon_points'] = bike_df.apply(lambda row: Point(latitude=row['start_station_latitude'], longitude=row['start_station_longitude']), axis=1)
bike_df['end_latlon_points'] = bike_df.apply(lambda row: Point(latitude=row['end_station_latitude'], longitude=row['end_station_longitude']), axis=1)
bike_df.head(3)
```

```python
#Then, calculate the KM between start and end points

bike_df['distance_in_km'] = bike_df.apply(lambda row: distance(row['start_latlon_points'], row['end_latlon_points']).km if row['end_latlon_points'] is not None else float('nan'), axis=1)
bike_df.head(3)
```

> Lets drop the un-used columns

```python
#reference: https://www.geeksforgeeks.org/copy-python-deep-copy-shallow-copy/
#the deepcopy function is way better when i try to work in heavy and big data

edit_bike_df = copy.deepcopy(bike_df)
edit_bike_df.head(2)
```

```python
edit_bike_df = edit_bike_df.drop(['start_time', 'end_time', 'start_station_id'
                                 , 'end_station_id', 'bike_id'
                                  , 'start_latlon_points' , 'end_latlon_points', 'start_station_name', 'end_station_name' ], axis = 1)
```

```python
edit_bike_df.head(3)
```

```python
edit_bike_df.shape
```

```python
edit_bike_df.columns
```

> Need to orgnize the columns little bit

```python
edit_bike_df = edit_bike_df[['user_type', 'bike_share_for_all_trip', 'member_birth_year', 'member_gender'
                             , 'month', 'weekday', 'hours', 'duration_sec',
                            'start_station_latitude', 'start_station_longitude',
                            'end_station_latitude', 'end_station_longitude',
                            'distance_in_km']]
```

```python
edit_bike_df.head(1)
```

```python
edit_bike_df.dtypes
```

```python
edit_bike_df[['member_birth_year', 'hours']] = edit_bike_df[['member_birth_year', 'hours']].astype(int)
edit_bike_df.dtypes
```

## Bivariate Exploration

>Lets do pair plotting for some data regards to :
    
    1- Sharing Bikes

Comparing between the shared_trips and the other columns with the highest values located in (NO) I think it will be unreasonable. So, I will take samples according to the lowest data on the Yes side, to plot the difference in the distance that the shared bike and not-shared one have took.

```python
#separate them into two DFs

yes_share_df = exp_bike_df[exp_bike_df['bike_share_for_all_trip'] == 'Yes']
no_share_df = exp_bike_df[exp_bike_df['bike_share_for_all_trip'] == 'No']
```

```python
#print shape to see the minimum rows
yes_share_df.shape
```

```python
#take 69k random samples for each dfs
yes_share_df = yes_share_df.sample(n=69000, replace = False, random_state=1)
no_share_df = no_share_df.sample(n=69000, replace = False, random_state=1)
```

```python
print(yes_share_df.shape)
print(no_share_df.shape)
```

```python
share_trip_df = pd.concat([yes_share_df, no_share_df], sort=False)
```

```python
#The two plots will be in vertical plotting visualization
#1st hours with km

plt.rcParams['font.size']=12
plt.rcParams['xtick.labelsize']=10
plt.rcParams['ytick.labelsize']=10

km_hr = sns.catplot(
    data=share_trip_df, kind="bar",
    x="hours", y="distance_in_km", hue="bike_share_for_all_trip",
      alpha=.6, height=6, ci=None)
km_hr.despine(left=True)
km_hr.fig.suptitle('Average kilometre vs. Average total hours from 4-months' , fontsize=15, y=1.05)
plt.show()

#2nd month with km

plt.rcParams['font.size']=12
plt.rcParams['xtick.labelsize']=10
plt.rcParams['ytick.labelsize']=10

km_m = sns.catplot(
    data=share_trip_df, kind="bar",
    x="month", y="distance_in_km", hue="bike_share_for_all_trip",
      alpha=.6, height=6, ci=None)
km_m.despine(left=True)
km_m.fig.suptitle('Average kilometre vs. the 4 Months' , fontsize=15, y=1)
plt.show()
```

> The shared vs. non-shared bike have been compared under total hours from 4 months ( which mean it took the average for hours starting from 00:00 to 23:00) and average distance in Kilometer

> the hours vs km figure shows that most of the standard hours were NOT shared except at 8:00, 14:00, and 17:00 o'clock. which mean during these three different hours the bike were shared.
            
>  While in Month vs Kilometer figure the bike was only shared in 4th month
